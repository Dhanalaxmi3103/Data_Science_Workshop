# *Handling Missing values*

```python
data_frame.isnull() #To check where null values are present
```

```python
data_frame.isnull().sum() #Gives number of null values in each atttribute
```

```python
#Filling null value in "Age" with mean of "Age"
data_frame["Age"].fillna(data_frame["Age"].mean(),inplace=True)
data_frame.isnull().sum()
```

```python
#Filling "Region" by randomly selecting region from "Region"
data_frame["Region"].fillna(np.random.choice(data_frame["Region"].unique()),inplace=True)
data_frame.isnull().sum()
```

```python
data_frame.dropna(inplace=True) #To remove all  rows having null values
data_frame.isnull().sum()
```

Missing values are checked, and different strategies are applied:

Age is filled with the mean of the column.
Region is filled randomly from the existing regions.
Rows with null values are also removed to ensure clean data for analysis.

```python
data_frame['Price']=data_frame['Total Price']+data_frame['Shipping Fee'] #Adding new column "Price" by adding "Total Price" and "Shipping Fee"
data_frame.head(5)
```

```python
data_frame.describe()#To know statistical information about dataset after handling missing values
```

A new column, Price, is created by summing the Total Price and Shipping Fee. This additional column gives a more complete view of the sales data.



```python
df=data_frame.drop(["Customer ID","Age"],axis=1)#Droping unrelated columns and loading it into another object
df.head(5)
```

Irrelevant columns like Customer ID and Age are dropped for further analysis, focusing only on useful columns such as product name, price, and quantity.


##**IDENTIFYING HIGH PERFORMING PRODUCTS**

```python
#Grouping by Product Name and calculating Total Price and Total Quantity Sold
products=df.groupby("Product Name").agg({"Price":"sum","Quantity":"sum"}).reset_index()
products
```

The code groups the data by Product Name, calculating the total sales price and total quantity sold for each product.



## *High Performing Products vs Price*

```python
#sorting Products by Total Price(descending)
high_performing=products.sort_values(by="Price",ascending=False)
high_performing
```

```python
#Bar chart of High performing Products by Total Price
plt.figure(figsize=(10,5))
plt.bar(high_performing["Product Name"],high_performing["Price"],color="m")
plt.title("High Performing Products")
plt.xlabel("Product Name")
plt.ylabel("Total Price")
plt.show()
```

## *High Performing Products vs Quantity*

```python
#Sorting Products by Quantities Sold(descending)
high_Quantity=products.sort_values(by="Quantity",ascending=False)
high_Quantity
```

```python
#Bar chart of High performing products by Quantities Sold
plt.figure(figsize=(10,5))
plt.bar(high_Quantity["Product Name"],high_Quantity["Quantity"],color="orange")
plt.title("High Performing Products")
plt.xlabel("Product Name")
plt.ylabel("Quantity")
plt.show()
```

A bar chart visualizes high-performing products based on total price and total quantity sold. The products with higher bars indicate better performance.
This analysis helps identify which products are driving the most revenue and which ones are sold in the highest quantities.


## *Product vs Shipping Status*

```python
#Grouping by Product Name and Shipping Status and Calculating Quantity
shipping_status=df.groupby(["Product Name","Shipping Status"]).agg({"Quantity":"sum"}).reset_index()
shipping_status
```

```python
#Plotting Product Name,Shipping Status and Quantity
for product in df["Product Name"].unique():
    product_data=shipping_status[shipping_status["Product Name"]==product]
    plt.subplots(figsize=(4,3))
    plt.bar(product_data["Shipping Status"],product_data["Quantity"],color="r")
    plt.title(f"Shipping Status of {product}")
    plt.xlabel("Shipping Status")
    plt.ylabel("Quantity")
    plt.show()
```

A deeper analysis is done on shipping status by product. This helps visualize which products are shipped successfully and identify any potential issues in product fulfillment.

## *Age vs Quantity of each Product*

```python
#Groupin by Product Name ,Age and Calculating Total Price and Quantity
age_group=data_frame.groupby(["Product Name","Age"]).agg({"Price":"sum","Quantity":"sum"}).reset_index()
age_group
```

```python
#Scatter Plot for Product Name,Age and Quantity
for product in df["Product Name"].unique():
    product_data=age_group[age_group["Product Name"]==product]
    plt.subplots(figsize=(4,3))
    plt.scatter(product_data["Age"],product_data["Quantity"],color="b")
    plt.title(f"Age vs Quantityof {product}")
    plt.xlabel("Age")
    plt.ylabel("Quantity")
    plt.show()
```

 A scatter plot is used to analyze the relationship between Age and Quantity Sold for each product. This could help identify target customer demographics for each product.

## **SEASONAL TREND ANALYSIS**

```python
#Exctracting month and year from "Order Date" column
df["Order Date"]=pd.to_datetime(df["Order Date"])
df["Month"]=df["Order Date"].dt.month
df["Year"]=df["Order Date"].dt.year
df.head(5)

#Grouping by year and month ,Then calculating Total Price for each month
monthly_sales=df.groupby(["Year","Month"]).agg({"Total Price":"sum"}).reset_index()
monthly_sales
```

The Order Date is converted into a DateTime object to easily extract Month and Year.

## *MONTHLY SALES TRENDS*

```python
#plotting Monthly Sales Trends
for year in df["Year"].unique():
    year_data=df[df["Year"]==year]
    monthly_sales=year_data.groupby("Month").agg({"Total Price":"sum"}).reset_index()
    plt.figure(figsize=(10,5))
    plt.plot(monthly_sales["Month"],monthly_sales["Total Price"],marker="d",mfc='m',mec='g',ms=10,color='c',label=f"Year {year}")
    plt.title("Monthly Sales Trends")
    plt.xlabel("Month")
    plt.ylabel("Total Price")
    plt.legend()
    plt.show()
```

 Monthly sales trends are analyzed by grouping the data by Year and Month and summing the total price for each month. A line plot is used to visualize how sales have fluctuated over time for each year.

## *SEASONAL SALES ANALYSIS*

```python
#function to determine season
def get_season(month):
    if month in [3,4,5]:
        return "Spring"
    elif month in [6,7,8]:
        return "Summer"
    elif month in [9,10,11]:
        return "Autumn"
    else:
        return "Winter"
#Apply function to create a new column for season
df["Season"]=df["Month"].apply(get_season)
print(df.head(5))
```

```python
#Grouping by seasons
seasons_group=df.groupby("Season").agg({"Price":"sum"}).reset_index()
print(seasons_group)

#plotting  seasonal  sales trends
plt.plot(seasons_group["Season"],seasons_group["Price"],marker="o",mec='r')
plt.title("Seasonal Sales Trends")
plt.xlabel("Seasons")
plt.ylabel("Total Price")
plt.show()
```

The data is categorized into four seasons (Spring, Summer, Autumn, and Winter) based on the month. Sales data is then aggregated by season to analyze the seasonal sales trends. A seasonal sales trend plot is created to show total sales per season.



```python
#Finding Total Price for each Product Name in all Seasons
monitor_sales=df[df["Product Name"]=="Monitor"].groupby(["Season"]).agg({"Price":"sum"}).reset_index()
print('monitor_sales \n')
print(monitor_sales)
keyboard_sales=df[df["Product Name"]=="Keyboard"].groupby(["Season"]).agg({"Price":"sum"}).reset_index()
print('keyboard_sales \n')
print(keyboard_sales)
mouse_sales=df[df["Product Name"]=="Mouse"].groupby(["Season"]).agg({"Price":"sum"}).reset_index()
print('mouse_sales \n')
print(mouse_sales)
smartphone_sales=df[df["Product Name"]=="Smartphone"].groupby(["Season"]).agg({"Price":"sum"}).reset_index()
print('smartphone_sales \n')
print(smartphone_sales)
smartwatch_sales=df[df["Product Name"]=="Smartwatch"].groupby(["Season"]).agg({"Price":"sum"}).reset_index()
print('smartwatch_sales \n')
print(smartwatch_sales)
laptop_sales=df[df["Product Name"]=="Laptop"].groupby(["Season"]).agg({"Price":"sum"}).reset_index()
print('laptop_sales \n')
print(laptop_sales)
headphones_sales=df[df["Product Name"]=="Headphones"].groupby(["Season"]).agg({"Price":"sum"}).reset_index()
print('headphones_sales \n')
print(headphones_sales)
```

The code analyzes the sales for individual products (like Monitor, Keyboard, Smartphone, etc.) in each season, allowing the identification of seasonal trends specific to each product. This is useful for seasonal marketing strategies

```python
#Plotting a stacked bar plot on seasonal sales
plt.figure(figsize=(10,5))
plt.bar(monitor_sales["Season"],monitor_sales["Price"],color="m",label="Monitor")
plt.bar(keyboard_sales["Season"],keyboard_sales["Price"],color="c",label="Keyboard",bottom=monitor_sales["Price"])
plt.bar(mouse_sales["Season"],mouse_sales["Price"],color="r",label="Mouse",bottom=monitor_sales["Price"]+keyboard_sales["Price"])
plt.bar(smartphone_sales["Season"],smartphone_sales["Price"],color="g",label="Smartphone",bottom=monitor_sales["Price"]+keyboard_sales["Price"]+mouse_sales["Price"])
plt.bar(smartwatch_sales["Season"],smartwatch_sales["Price"],color="y",label="Smartwatch",bottom=monitor_sales["Price"]+keyboard_sales["Price"]+mouse_sales["Price"]+smartphone_sales["Price"])
plt.bar(laptop_sales["Season"],laptop_sales["Price"],color="k",label="Laptop",bottom=monitor_sales["Price"]+keyboard_sales["Price"]+mouse_sales["Price"]+smartphone_sales["Price"]+smartwatch_sales["Price"])
plt.bar(headphones_sales["Season"],headphones_sales["Price"],color="b",label="Headphones",bottom=monitor_sales["Price"]+keyboard_sales["Price"]+mouse_sales["Price"]+smartphone_sales["Price"]+smartwatch_sales["Price"]+laptop_sales["Price"])
plt.title("Seasonal Sales Trends")
plt.xlabel("Seasons")
plt.ylabel("Total Price")
plt.legend()
plt.show()
```

 A stacked bar plot is used to compare product sales in different seasons. This visualization shows how sales for different products contribute to total sales in each season.
