# Visualize the distribution of a categorical column

```python
#category column
df.select_dtypes(include='object').columns
```

```python
df.workclass.value_counts()
```

```python
#replace ? with unknown
df.workclass.replace('?', 'unknown', inplace=True)
```

```python
#plot of workclass
df.workclass.value_counts().plot(kind='bar')
plt.xlabel('workclass')
plt.ylabel('count')
plt.title('workclass distribution')
plt.show()
```

```python
df.education.value_counts(normalize=True).plot.bar()
plt.xlabel('education')
plt.ylabel('count')
plt.title('education distribution')
plt.show()
```

```python
df.groupby('marital.status').size().plot.bar()
plt.xlabel('marital.status')
plt.ylabel('count')
plt.title('marital.status distribution')
plt.show()
```

```python
df.sex.value_counts().plot.bar()
plt.xlabel('sex')
plt.ylabel('count')
plt.title('sex distribution')
plt.show()
```

```python
df.occupation.value_counts().plot.bar()
plt.xlabel('occupation')
plt.ylabel('count')
plt.title('occupation distribution')
plt.show()
```

```python
df.relationship.value_counts().plot.pie()
plt.title('relationship distribution')
plt.show()
```

# Visualize the relationship between two numerical columns using a scatter plot.

```python
plt.scatter(df['age'],df['fnlwgt'],color='y',marker='*')
plt.xlabel('age')
plt.ylabel('fnlwgt')
plt.title('age vs fnlwgt')
plt.show()
```

```python
plt.scatter(df['age'],df['hours.per.week'],color='orange')
plt.xlabel('age')
plt.ylabel('hours.per.week')
plt.title('age vs hours.per.week')
plt.show()
```

```python
plt.scatter(df['fnlwgt'],df['hours.per.week'],color='c',marker='d')
plt.xlabel('fnlwgt')
plt.ylabel('hours.per.week')
plt.title('fnlwgt vs hours.per.week')
plt.show()
```

## numeric vs category

```python
df.groupby('workclass')['age'].mean().plot.bar()
plt.xlabel('workclass')
plt.ylabel('age')
plt.title('workclass vs age')
plt.show()
```

```python
df.groupby('education')['age'].sum().plot.bar()
plt.xlabel('education')
plt.ylabel('age')
plt.title('education vs age')
plt.show()
```

```python
df.groupby('occupation')['hours.per.week'].sum().plot.bar()
plt.xlabel('occupation')
plt.ylabel('hours.per.week')
plt.title('occupation vs hours.per.week')
plt.show()
```

```python
df.groupby('race')['age'].sum().plot.barh(color='m')
plt.xlabel('age')
plt.ylabel('race')
plt.title('race vs age')
plt.show()
```
